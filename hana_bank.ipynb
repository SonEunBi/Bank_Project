```python
import pandas as pd
import numpy as np
import pyproj

df = pd.read_csv("2022_수정.csv", encoding="cp949",
                 usecols = ['점번호','gcrm은행코드', '한글점명',
                          '주소','가로위치좌표', '세로위치좌표'  ])
df.dropna()

df['가로위치좌표'] = pd.to_numeric(df['가로위치좌표'], errors="coerce")
df['세로위치좌표'] = pd.to_numeric(df['세로위치좌표'], errors="coerce")

df = df.dropna()
df.index=range(len(df))
df.tail()

# 좌표계 변환 함수
def project_array(coord, p1_type, p2_type):
    p1 = pyproj.Proj(init=p1_type)
    p2 = pyproj.Proj(init=p2_type)
    fx, fy = pyproj.transform(p1, p2, coord[:, 0], coord[:, 1])
    return np.dstack([fx, fy])[0]

coord = np.array(df)
coord

# 좌표계 정보 설정
p1_type = "epsg:2097"
p2_type = "epsg:4326"

# project_array() 함수 실행
result = project_array(coord, p1_type, p2_type)
result

print(result)

df['경도'] = result[:, 0]
df['위도'] = result[:, 1]

seoul = df[df['주소'].str.contains('서울특별시')]
ic = df[df['주소'].str.contains('인천광역시')] 
gg = df[df['주소'].str.contains('경기도')] 

display(seoul)
```
